```python
#linear regression
from sklearn.model_selection import train_test_split
```

```python
#converting string data to data-time
import datetime as dt
india_case['date']=pd.to_datetime(india_case['date'])
india_case.head()
```

```python
#converting date-time to ordinal
india_case['date']=india_case['date'].map(dt.datetime.toordinal)
india_case.head()
```

```python
#getting dependent variable and indipendent variable
import numpy as np
x=np.array(india_case['date']).reshape(-1,1)
y=np.array(india_case['total_cases']).reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
import numpy as np
lr.fit(x_train,y_train)
```

```python
india_case.tail()
```

```python
y_pred=lr.predict(np.array(x_test).reshape(-1,1))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(x_test,y_pred)
```
